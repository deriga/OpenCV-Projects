```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

print(cv.__version__)
```

```python
img1 = cv.imread("C:/opencv/opencv/medias/classic.jpg")
img2 = cv.imread("C:/opencv/opencv/medias/sport_car.jpg")

row, col = 1, 2
figure, axis = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()
 
axis[0].imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
axis[0].set_title('Classic')

axis[1].imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
axis[1].set_title('Sport Car')

plt.show()
```

## Black & White Image

```python
imgGray1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
imgGray2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)

row, col = 1, 2
figure, axis = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()
 
axis[0].imshow(cv.cvtColor(imgGray1, cv.COLOR_BGR2RGB))
axis[0].set_title('Black & White Classic')

axis[1].imshow(cv.cvtColor(imgGray2, cv.COLOR_BGR2RGB))
axis[1].set_title('Black & White Sport Car')

plt.show()
```

## Blur Image

```python
kernel = np.ones((5,5), np.uint8)

imgBlur1 = cv.GaussianBlur(imgGray1, (9,9),0)
imgBlur2 = cv.GaussianBlur(imgGray2, (9,9),0)

row, col = 1, 2
figure, axis = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()
 
axis[0].imshow(cv.cvtColor(imgBlur1, cv.COLOR_BGR2RGB))
axis[0].set_title('Blurred Classic')

axis[1].imshow(cv.cvtColor(imgBlur2, cv.COLOR_BGR2RGB))
axis[1].set_title('Blurred Sport Car')

plt.show()
```

## Canny Image

```python
imgCanny1 = cv.Canny(img1,200,100)
imgCanny2 = cv.Canny(img1,500,100) # Higher Threshold value of first image
imgCanny3 = cv.Canny(img2,200,100) 
imgCanny4 = cv.Canny(img2,500,100) # Higher Threshold value of second image

row, col = 2, 2
figure, axis = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()
 
axis[0][0].imshow(cv.cvtColor(imgCanny1, cv.COLOR_BGR2RGB))
axis[0][0].set_title('Canny Classic Low Threshold')

axis[0][1].imshow(cv.cvtColor(imgCanny2, cv.COLOR_BGR2RGB))
axis[0][1].set_title('Canny Classic High Threshold')

axis[1][0].imshow(cv.cvtColor(imgCanny3, cv.COLOR_BGR2RGB))
axis[1][0].set_title('Canny Sport Car Low Threshold')

axis[1][1].imshow(cv.cvtColor(imgCanny4, cv.COLOR_BGR2RGB))
axis[1][1].set_title('Canny Sport Car High Threshold')

plt.show()
```

## Dilated Image

```python
imgDilated1 = cv.dilate(imgCanny1, kernel, iterations=2)
imgDilated2 = cv.dilate(imgCanny2, kernel, iterations=1)
imgDilated3 = cv.dilate(imgCanny3, kernel, iterations=3)
imgDilated4 = cv.dilate(imgCanny4, kernel, iterations=1)
kernel = np.ones((5,5), np.uint8)


row, col = 2, 2
figure, axis = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()
 
axis[0][0].imshow(cv.cvtColor(imgDilated1, cv.COLOR_BGR2RGB))
axis[0][0].set_title("Dilated Classic")

axis[0][1].imshow(cv.cvtColor(imgDilated2, cv.COLOR_BGR2RGB))
axis[0][1].set_title("Dilated Classic")

axis[1][0].imshow(cv.cvtColor(imgDilated3, cv.COLOR_BGR2RGB))
axis[1][0].set_title("Dilated Sport Car")

axis[1][1].imshow(cv.cvtColor(imgDilated4, cv.COLOR_BGR2RGB))
axis[1][1].set_title("Dilated Sport Car")

plt.show()
```

## Eroded Image

```python
imgEroded1 = cv.erode(imgDilated1, kernel, iterations=1)
imgEroded2 = cv.erode(imgDilated2, kernel, iterations=1)
imgEroded3 = cv.erode(imgDilated3, kernel, iterations=1)
imgEroded4 = cv.erode(imgDilated4, kernel, iterations=1)
kernel = np.ones((5,5), np.uint8)

row, col = 2, 2
figure, axis = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()

axis[0][0].imshow(cv.cvtColor(imgEroded1, cv.COLOR_BGR2RGB))
axis[0][0].set_title("Eroded Classic")

axis[0][1].imshow(cv.cvtColor(imgEroded2, cv.COLOR_BGR2RGB))
axis[0][1].set_title("Eroded Classic")

axis[1][0].imshow(cv.cvtColor(imgEroded3, cv.COLOR_BGR2RGB))
axis[1][0].set_title("Eroded Sport Car")

axis[1][1].imshow(cv.cvtColor(imgEroded4, cv.COLOR_BGR2RGB))
axis[1][1].set_title("Eroded Sport Car")

plt.show()
```
