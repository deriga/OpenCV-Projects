```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

print(cv.__version__)
```

```python
# Create two black image
image = np.zeros((512,512,3), np.uint8)
image_2 = np.zeros((512,512,3), np.uint8)

row, col = 1, 2
figure, axis = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()
 
axis[0].imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
axis[0].set_title('black image 1')

axis[1].imshow(cv.cvtColor(image_2, cv.COLOR_BGR2RGB))
axis[1].set_title('black image 2')

plt.show()
```

```python
# Draw a diagonal blue line of thickness of 5 pixels
cv.line(image, (0,0), (512,512), (255,0,0), 50)

#Draw a cross within square with cv.line
cv.line(image_2,(100,100),(100,412),(0,255,0),20)
cv.line(image_2,(100,412),(412,412),(0,255,0),20)
cv.line(image_2,(412,100),(412,412),(0,255,0),20)
cv.line(image_2,(100,100),(412,100),(0,255,0),20)
cv.line(image_2,(115,115),(397,397),(0,0,255),10)
cv.line(image_2,(115,397),(397,115),(0,0,255),10)

row, col = 1, 2
figure, image_num = plt.subplots(row, col, figsize=(15, 10))
figure.tight_layout()

image_num[0].imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
image_num[0].set_title('Cross linear line')

image_num[1].imshow(cv.cvtColor(image_2, cv.COLOR_BGR2RGB))
image_num[1].set_title('Red cross within green square')

plt.show()
```

```python
image_rectangle = np.zeros((512,512,3),np.uint8)

cv.rectangle(image_rectangle, (100,100), (412,412), (0,255,255), 50)
cv.rectangle(image_rectangle, (200,200), (312,312),(0,0,255), 20)

plt.imshow(cv.cvtColor(image_rectangle, cv.COLOR_BGR2RGB))
plt.title("Rectangles")

plt.show()
```

```python
image_circle = np.zeros((512,512,3), np.uint8)

cv.circle(image_circle, (110,110), 80, (255,0,0), 50)
cv.circle(image_circle, (320,320), 180, (255,255,0),20)
cv.circle(image_circle, (320,320), 160, (0,255,0), -1)

plt.imshow(cv.cvtColor(image_circle, cv.COLOR_BGR2RGB))
plt.title("Circles")

plt.show()
```

```python
image_hexagon = np.zeros((512,512,3), np.uint8)

points = np.array([[50,256], [150,429], [350,429], [450,256], [350,82], [150,82]], np.int32)

points = points.reshape((-1,1,2))

cv.polylines(image_hexagon, [points], True, (0,0,255), 30)

plt.imshow(cv.cvtColor(image_hexagon, cv.COLOR_BGR2RGB))
plt.title('Hexagon')

plt.show()
```

```python
text= 'This is a Hexagon'

cv.putText(image_hexagon, text, (120,256), cv.FONT_HERSHEY_COMPLEX, 0.8, (127,255,255), 2)

plt.imshow(cv.cvtColor(image_hexagon, cv.COLOR_BGR2RGB))
plt.title('Hexagon with text')
plt.show()
```
